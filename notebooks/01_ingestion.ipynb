```python
from minio import Minio
import os

client = Minio(
    "minio:9000",
    access_key="minioadmin",
    secret_key="minioadmin",
    secure=False
)

BUCKET = "bigdata"
LOCAL_FILE = "/home/jovyan/data/raw/Retail_Transactions_Dataset.csv" 
OBJECT_NAME = "bronze/retail_transactions/retail_transactions.csv"

if not client.bucket_exists(BUCKET):
    client.make_bucket(BUCKET)
    print(f"Bucket '{BUCKET}' dibuat.")
else:
    print(f"Bucket '{BUCKET}' sudah ada.")

client.fput_object(
    BUCKET,
    OBJECT_NAME,
    LOCAL_FILE,
    content_type="text/csv"
)
print(f"File berhasil diupload ke s3a://{BUCKET}/{OBJECT_NAME}")
```

```python
objects = client.list_objects(BUCKET, prefix="bronze/", recursive=True)
print("Objek di Bronze Layer:")
for obj in objects:
    print(f"  - {obj.object_name}  ({obj.size} bytes)")
```
